## Define

Remove newline characters in Team & Contracts and Hits columns 

## Code

```python
df.Hits.describe()
```

```python
#remove newline character from Hits column
df.Hits = df.Hits.str.strip('\n')

#creating function to convert Hits values with K
def change_hit(x):
    if 'K' in str(x):
        x = float(x.strip('K')) * 1000
        return x
    else:
        x = float(x)
        return x
df.Hits = df.Hits.apply(change_hit)
```

```python
#Categorizing players based on their contract type
def categorize(x):
    if 'Free' in x:
        x = 'Free'
        return x
    elif 'Loan' in x:
        x = 'On Loan'
        return x
    else:
        x = 'Contract'
        return x
df['Contract Type'] = df['Team & Contract'].apply(categorize)
```

```python
#number of Players in the 3 types of contracts
df['Contract Type'].value_counts()
```

```python
#getting the club name of each player in fifa 21
df['Club'] = df['Team & Contract'].apply(lambda x: x.strip('\n').split('\n')[0])
```

```python
#getting the contract duration year of each player
df['Contract'] = df['Team & Contract'].apply(lambda x: np.nan if ("Loan" in x or "Free" in x) else x.strip('\n').split('\n')[1])
df['Contract Year Begins'] = df['Contract'].str[0:4]
df['Contract Year Ends'] = df['Contract'].str[-4:]
```

```python
#removing unecessary columns
df.drop(['Team & Contract', 'Contract'], axis=1, inplace=True)
```

## Define

Remove characters and convert values in Value, Wage and Release Clause 

## Code

```python
#creating a function that removes pound symbol, corrects datatype and converts each value for the Value, Wage and Release clause columns
def money(x):
    if 'M' in x:
        x = float(x.strip('€').strip('M')) * 1000000
        return x
    elif 'K' in x:
        x = float(x.strip('€').strip('K')) * 1000
        return x
    else:
        x = float(x.strip('€'))
        return x

df['Value'] = df['Value'].apply(money)
df['Wage'] = df['Wage'].apply(money)
df['Release Clause'] = df['Release Clause'].apply(money)
```

## Define

convert height and weight to appropraite value

## Code

```python
#5'7" means 5 feets 7 inches
# to convert feet to cm multiply the value by 30.48
# to convert inches to cm multiply the value by 2.54
```

```python
#Creating a function that convert Height and Weight
def convert(x):
    if "'" in x:
        f = int(x.split("'")[0]) * 30.48
        i = int(x.split("'")[1][0]) * 2.54
        x = f + i
        return x
    elif "lbs" in x:
        x = round(int(x.strip('lbs')) / 2.205, 2)
        return x

df.Height = df.Height.apply(convert)
df.Weight = df.Weight.apply(convert)
```
